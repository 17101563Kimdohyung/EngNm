# X3DOM Examples

This notebook demonstrates experimental DOLFIN features (http://fenicsproject.org) for X3DOM (http://www.x3dom.org/) and HTML output. The objective is to support the interactive display of meshes and simulation outputs in web browsers, including in Jupyter notebooks.

```python
from IPython.core.display import HTML
from dolfin import *
import mshr
```

```python
# Create mesh in 2D and display mesh
mesh = UnitSquareMesh(32, 32)
mesh
#HTML(X3DOM.html(mesh))
```

```python
# Create mesh in 3D and display mesh
mesh = UnitCubeMesh(16, 16, 16)
mesh
#HTML(X3DOM.html(mesh))
```

```python
c = mshr.Circle(Point(0, 0), 1.5)
e = mshr.Ellipse(Point(0, 0), .5, .9)

n_cells = 20
mesh = mshr.generate_mesh(c -e, n_cells)
mesh
#HTML(X3DOM.html(mesh))
```

---

```python
# Create a contour plot
e = Expression("sin(x[0])*sin(x[1])", degree=1)
V = FunctionSpace(mesh, "Lagrange", 1)
u = Function(V)
u.interpolate(e)
```

```python
HTML(X3DOM.html(u.cpp_object()))
```

```python
# Create mesh and define function space
mesh = UnitCubeMesh(16, 16, 16)
V = FunctionSpace(mesh, "Lagrange", 1)
```

```python
# Define Dirichlet boundary at the bottom face
def boundary(x):
    return x[0] < DOLFIN_EPS 

# Define boundary condition
bc = DirichletBC(V, Constant(0.0), boundary)
```

```python
# Define variational problem
u = TrialFunction(V)
v = TestFunction(V)
f = Expression("10*exp(-(pow(x[0] -0.5, 2) + pow(x[1] -0.5, 2) + pow(x[2] -0.5, 2)) /0.02)", degree=1)
a = inner(grad(u), grad(v))*dx
L = f*v*dx 

# Compute solution
u = Function(V)
solve(a == L, u, bc)
```

```python
HTML(X3DOM.html(u.cpp_object()))
```
