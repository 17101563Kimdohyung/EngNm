### 1. Write a Python program to create stack bar plot and add label to each section.

> Sample data:

```python
people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')

# multi-dimensional data 
data = [[  3.40022085, 7.70632498,  6.4097905,  10.51648577,  7.5330039,   7.1123587,  12.77792868, 3.44773477],
        [ 11.24811149, 5.03778215,  6.65808464, 12.32220677,  7.45964195,  6.79685302,  7.24578743, 3.69371847],
        [ 3.94253354,  4.74763549, 11.73529246,  4.6465543,  12.9952182,   4.63832778, 11.16849999, 8.56883433],
        [ 4.24409799, 12.71746612, 11.3772169,   9.00514257, 10.47084185, 10.97567589,  3.98287652, 8.80552122]]
```

> The code snippet gives the output shown in the following screenshot:

````python
import numpy as np
import matplotlib.pyplot as plt

people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')

# multi-dimensional data 
data = [[  3.40022085, 7.70632498,  6.4097905,  10.51648577,  7.5330039,   7.1123587,  12.77792868, 3.44773477],
        [ 11.24811149, 5.03778215,  6.65808464, 12.32220677,  7.45964195,  6.79685302,  7.24578743, 3.69371847],
        [ 3.94253354,  4.74763549, 11.73529246,  4.6465543,  12.9952182,   4.63832778, 11.16849999, 8.56883433],
        [ 4.24409799, 12.71746612, 11.3772169,   9.00514257, 10.47084185, 10.97567589,  3.98287652, 8.80552122]]

y_pos = np.arange(len(people))
percentages = np.array(data) /np.sum(data, axis=0) *100

fig, ax = plt.subplots(figsize=(10, 6))

colors ='rgbm'
patch_handles = []
# left alignment of data starts at zero
left = np.zeros(len(people)) 
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d, color=colors[i%len(colors)], align='center', left=left))
    left += d

# search all of the bar segments and annotate
for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5 *patch.get_width() +bl[0]
        y = 0.5 *patch.get_height() +bl[1]
        ax.text(x, y, "%.1f%%" % (percentages[j, i]), ha='center')

ax.tick_params(direction='in')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlim([0, 40])
ax.set_xlabel('Scores')

plt.savefig('ex01_matplotlib.png');
````

### 2. Write a Python programming to create a pie chart with a title of the popularity of programming languages. 

```
Sample data:
    Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8.0, 7.7, 6.7
```

> The code snippet gives the output shown in the following screenshot:

````python
import matplotlib.pyplot as plt

# Plot data
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

#colors = ['red', 'gold', 'yellowgreen', 'blue', 'lightcoral', 'lightskyblue']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]

# explode 1st slice
explode = (0.1, 0, 0, 0, 0, 0)  

fig, ax = plt.subplots(figsize=(8, 8))

# Plot
ax.pie(popuratity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Popularity of Programming Language\n Worldwide, Oct. 2017", bbox={'facecolor':'0.8', 'pad':5})

plt.savefig('ex02_matplotlib.png');
````
